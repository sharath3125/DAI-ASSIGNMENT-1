## Load the Dataset

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Check for Missing Values

```python
# Load the dataset
automobile_data = pd.read_csv('file:///Users/vivekrathod/Downloads/Automobile.csv')
```

```python
# ## Step 1: Check for Missing Values
# Let's identify any missing values in the dataset.
missing_values = automobile_data.isnull().sum()
print("### Missing Values:\n", missing_values)
```

```python
# ## Step 2: Handle Missing Values
# Since we found missing values in the 'horsepower' column, we'll fill these with the column's mean.
automobile_data['horsepower'].fillna(automobile_data['horsepower'].mean(), inplace=True)
```

## Handle Missing Values

```python
# ## Step 3: Check for Outliers in Numerical Columns
# Generate summary statistics to observe the range and distribution of each numeric column.
numeric_cols = automobile_data.select_dtypes(include=['float64', 'int64']).columns
numeric_summary = automobile_data[numeric_cols].describe()
print("### Summary Statistics:\n", numeric_summary)
```

## Summary Statistics and Outlier Detection

```python
# ## Step 4: Univariate Analysis - Histograms for Numeric Columns
# We’ll plot histograms for each numeric column to observe their distributions.
for column in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(automobile_data[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

## Univariate Analysis

```python
# ## Step 5: Bivariate Analysis - Pairplot to Observe Relationships
# Use pairplot to explore relationships between numeric variables.
sns.pairplot(automobile_data[numeric_cols])
plt.suptitle('Pairplot of Numeric Columns', y=1.02)
plt.show()
```

## Bivariate Analysis - Pairplot
